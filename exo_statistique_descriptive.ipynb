## Question 13:L'asymetrie des variable numerique

````python
from scipy import stats
numerique_data = df.select_dtypes(include=[np.number])
for charge in numerique_data.columns:
    stat,p_value = stats.skewtest(numerique_data[charge])
    if p_value < 0:
        print(f"Les données de {charge} sont asymétriques,distribution décalée à droite de la médiane avec un p-value de {p_value} qui inferieur à 0.05  et un skew de {stat}")
    elif p_value > 0:  
        print(f"Les données de {charge} sont asymétriques,distribution décalée à guache de la médiane avec un p-value de {p_value} qui inferieur à 0.05  et un skew de {stat}")
      
    else:
        print(f"Les données de {charge} sont symétriques avec un p-value de {p_value} qui superieur à 0.05 et un skew de {stat}")
````

## Question 13:Kurtosis des variables numeriques

````python
for charge in numerique_data.columns:
    kurt = stats.kurtosis(numerique_data[charge])
    if kurt > 0:
        print(f"Les données de {charge} sont leptokurtique avec un kurtosis de {kurt}")
    elif kurt < 0:
        print(f"Les données de {charge} sont platykurtique avec un kurtosis de {kurt}")
    else:
        print(f"Les données de {charge} sont mesokurtique avec un kurtosis de {kurt}")
````

## question 9:Visualize the correlations between the SalePrice variable and all other variable
