```python
data_dir = 'E:/Data/script/Datasets/rnn_dataset/victorhugo.txt'
```

```python
with open(data_dir,'r',encoding="utf-8") as f:
    text = f.read()

print(text)
```

```python
replace_caracter = ['2','1','8','5','<','>','!','?','-','$',';']

import unidecode

text = unidecode.unidecode(text)
text = text.lower()
text.strip()
for caract in replace_caracter:
    text = text.replace(caract ,"")
```

```python
vocab = set(text)
```

```python
print(text[:100])
```

```python
vocab_to_int = {j:i for i,j in enumerate(vocab)}
int_to_vocab = {i:j for i,j in enumerate(vocab)}
```

```python
encoded = [vocab_to_int[i] for i in text]
print(encoded[:100])
```

```python
decoded = [int_to_vocab[i] for i in encoded]
```

```python
print(''.join(decoded[:100]))
```
